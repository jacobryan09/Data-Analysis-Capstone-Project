## Question 1: How fast is ChatGPT growing over time? 

```python
# Calculate Monthly Growth Rate (%)
merged_usage['visit_growth_%'] = merged_usage['monthly_visits_(millions)'].pct_change() * 100
merged_usage['user_growth_%'] = merged_usage['users_(millions)'].pct_change() * 100

# Display key growth metrics
merged_usage[['month', 'monthly_visits_(millions)', 'visit_growth_%', 'users_(millions)', 'user_growth_%']]
```

### Visualize the Data

```python
import matplotlib.pyplot as plt

# Line chart for Monthly Visits
plt.figure(figsize=(10, 5))
plt.plot(merged_usage['month'], merged_usage['monthly_visits_(millions)'], marker='o')
plt.title('Monthly Visits to ChatGPT (Millions)')
plt.xlabel('Month')
plt.ylabel('Visits (Millions)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Line chart for Total Users
plt.figure(figsize=(10, 5))
plt.plot(merged_usage['month'], merged_usage['users_(millions)'], color='green', marker='o')
plt.title('Total ChatGPT Users Over Time (Millions)')
plt.xlabel('Month')
plt.ylabel('Users (Millions)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Growth Rate Chart
plt.figure(figsize=(10, 5))
plt.plot(merged_usage['month'], merged_usage['visit_growth_%'], label='Visit Growth %', marker='o')
plt.plot(merged_usage['month'], merged_usage['user_growth_%'], label='User Growth %', marker='o')
plt.title('Monthly Growth Rate of Visits and Users')
plt.xlabel('Month')
plt.ylabel('Growth %')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Predicting Future ChatGPT Users (Next 12 Months)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Assume 'data' is already cleaned and loaded with 'month' and 'total_users'
data['month'] = pd.to_datetime(data['month'])
data = data.sort_values('month')

# Create time index
data['time_index'] = np.arange(len(data))

# Train model
X = data[['time_index']]
y = data['total_users']
model = LinearRegression()
model.fit(X, y)

# Predict on training data
data['predicted_users'] = model.predict(X)

# Forecast next 10 years (120 months)
future_time_index = pd.DataFrame({'time_index': np.arange(len(data), len(data) + 120)})
future_predictions = model.predict(future_time_index)

# Create future months
last_date = data['month'].max()
future_months = pd.date_range(start=last_date + pd.DateOffset(months=1), periods=120, freq='MS')

# Forecast DataFrame
forecast_df = pd.DataFrame({
    'month': future_months,
    'predicted_users': future_predictions
})

# Plot
plt.figure(figsize=(12, 6))
plt.plot(data['month'], data['total_users'], 'o-', label='Actual Users')
plt.plot(data['month'], data['predicted_users'], 'r--', label='Predicted (Trained)')
plt.plot(forecast_df['month'], forecast_df['predicted_users'], 'g--', label='Forecast (2035)')
plt.title("ChatGPT User Growth Forecast Until 2035")
plt.xlabel("Month")
plt.ylabel("Users (millions)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Preview forecast
print(forecast_df.head(12))
```

### In this section of the project, a linear regression model was developed to forecast the future growth of ChatGPT users based on historical data from January 2023 to April 2025. The model predicts that the number of users will continue increasing at a steady rate, reaching approximately 518 million users by April 2026. The forecast assumes a consistent upward trend without factoring in possible market saturation, regulation, or user decline. While this projection is based on a simplified linear model, it provides a clear and interpretable outlook that supports the general expectation of continued AI adoption. For the purposes of this capstone project, the prediction is reasonable and supports the broader narrative that interest and usage of generative AI platforms like ChatGPT will likely keep expanding in the near future.

# III. Job-Threat Index

## This data shows what jobs that could be possibly replaced by AI in the upcoming years

### Top 5 Jobs that are in danger of AI replacement

```python
# Import Libraries
import pandas as pd

# Load the CSV file
data = pd.read_csv("job_threat_index.csv")

# Fix column name typo (from 'Job titles' to 'Job_titles')
data = data.rename(columns={'Job titiles': 'Job_titles'})

# Convert 'AI Impact' from percentage string to float
data['AI_Impact_Score'] = data['AI Impact'].str.replace('%', '').astype(float)

# Filter out rows where 'AI models' = 0 (to avoid division by zero)
data = data[data['AI models'] != 0]

# Recalculate workload ratio to be sure it's accurate
data['AI_Workload_Ratio'] = data['AI models'] / data['Tasks']

# Sort by AI Impact Score (or use AI_Workload_Ratio if preferred)
top_jobs = data.sort_values(by='AI_Impact_Score', ascending=False).head(5)

# Print top 5 jobs with index starting at 1
for idx, row in enumerate(top_jobs.itertuples(), start=1):
    print(f"{idx}. {getattr(row, 'Job_titles')} - AI Impact: {row.AI_Impact_Score}%, Workload Ratio: {row.AI_Workload_Ratio:.3f}")
```

### Top 5 Jobs that are least replaced by AI 

```python
# Sort by lowest AI Impact Score
least_replaceable_jobs = data.sort_values(by='AI_Impact_Score', ascending=True).head(5)

# Print the 5 least replaceable jobs (index starts at 1)
print("\nLeast Replaceable Jobs by AI:\n")
for idx, row in enumerate(least_replaceable_jobs.itertuples(), start=1):
    print(f"{idx}. {getattr(row, 'Job_titles')} - AI Impact: {row.AI_Impact_Score}%, Workload Ratio: {row.AI_Workload_Ratio:.3f}")
```

#### Industries that are most impacted by AI

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the cleaned CSV with normalized values
file_path = 'Normalized_Average_AI_Impact_by_Domain.csv'
data = pd.read_csv(file_path)

# Step 2: Rename column to match expected code
data = data.rename(columns={'AI_Impact_Clean': 'AI_Impact_Percent'})

# Step 3: Sort values for cleaner pie
data_sorted = data.sort_values(by='AI_Impact_Percent', ascending=False)

# Step 4: Plot pie chart
plt.figure(figsize=(10, 8))
plt.pie(
    data_sorted['AI_Impact_Percent'],
    labels=data_sorted['Domain'],
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.cm.Pastel1.colors  # soft, non-dark colors
)

plt.title('Normalized Average AI Impact by Industry Domain (Sum = 100%)')
plt.axis('equal')  # Makes sure pie is a circle
plt.tight_layout()
plt.show()
```

#### The pie chart shows that Administrative & Clerical jobs are the most affected by AI, followed closely by Data & IT, Sales & Marketing, and Communication & PR.
